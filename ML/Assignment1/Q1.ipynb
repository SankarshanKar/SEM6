```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([2017, 2018, 2019, 2020, 2021, 2022])
y = np.array([3000, 8000, 20000, 25000, 32000, 40000])
n = len(x)
```

```python
b1 = ((n*(np.sum(np.matmul(x,y))) - (np.sum(x)*np.sum(y))) / (((n*np.sum(np.matmul(x,x)))) - pow((np.sum(x)),2)))
b0 = (np.sum(y) - b1*(np.sum(x)))/n
y_pred = b0 + b1 * x
```

```python
plt.scatter (x, y, c = "red", label = "Actual Data")
plt.plot (x, y_pred, label = "Predicted Data")
plt.xlabel("Year")
plt.ylabel("Nexon Units Sold")
plt.legend ()
plt. show()
```

```python
height = np.array([5, 5.11, 5.6, 5.9, 4.8, 5.8, 5.5, 5.6])
age = np.array([45, 26, 30, 34, 40, 36, 19, 28, 23, 32])
weight = np.array([77, 47, 55, 59, 72, 60, 40, 60, 45, 58])
K = 5
n = len(age)
dist = np.zeros (n)
```

```python
for i in range(n):
    dist[i] = np.sqrt((38 - age[i]) * 2 + (5.6 - height[i]) * 2)
```

```python
index = np.argsort(dist)
print("The predicted weight is: ", np.round(np.mean(weight[index[:K]])))
```
